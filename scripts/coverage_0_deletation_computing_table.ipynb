```python

import pandas as pd
import numpy as np
import csv,re
import os,io,sys
```

```python


df1=pd.read_csv("/home/azadeh/workspace/snakemake_tutorial/snakemake-snakemake-tutorial-data-f17b467/coverage_by_base",delimiter="\t", header=None)
df1["longueur_gène"]=df1[2]-df1[1]
df1_genes=df1[[1,2,3,4,5,"longueur_gène"]]
df1_genes=df1_genes.rename(columns={ 1: 'début' , 2 : 'fin', 5 : 'profondueur', 3 : 'gène', 4:'base_position'})

df1_genes.tail(500)
```

```python
df1_genes=df1_genes[df1_genes.profondueur==0]
df1_genes["position_génomique"]=df1_genes.apply(lambda x : x[0]+x[3], axis=1)
df1_genes
```

```python
table=df1_genes.groupby(['gène', 'début', 'fin' , 'longueur_gène' ])
new_table=table.agg({'base_position' : [np.min, np.max , np.ptp] , 'position_génomique':[np.min]})

new_table
```
